```python
import sys
sys.path.append('../')
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.axes import Axes
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_csv('latencies.csv')
data["latency"] = data['end'] - data['start']
data.groupby("type")["latency"].describe()
```

```python
plt.figure(figsize=(7,4))
ax: Axes = sns.histplot(data,
    x=data["latency"]*1000,
    hue=data["type"].replace({
        "cameraonly": "Camera only",
        "delay": "Inertial data delayed",
        "prediction": "Inertial prediction (proposed)"}),
    stat="percent",
    common_norm=False,
    element="step",
    fill=True,
    binwidth=5,
    kde=True)
ax.set_xlabel("Motion-to photon latency (ms)")
ax.set(ylabel="% of measurements")
leg = ax.get_legend()
leg.set_title("")
leg.set_loc('upper right')
plt.savefig('latency-histogram.pdf', dpi=300)
```

```python
plt.figure(figsize=(6,4))
ax: Axes = sns.boxplot(data=data,
    y=data["type"].replace({
        "cameraonly": "Camera only",
        "delay": "Inertial data\ndelayed",
        "prediction": "Proposed"}),
    x=data["latency"] * 1000,
    )
ax.set(xlim=(0,150))
ax.set_xlabel("Motion-to-photon latency (ms). Lower is better.")
ax.set_ylabel("")
plt.tight_layout()
plt.savefig('latency-boxplot.png', dpi=300)
```
